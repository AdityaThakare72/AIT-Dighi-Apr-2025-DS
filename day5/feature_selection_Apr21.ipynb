```python
# data = paribas.csv
```

```python
df = pd.read_csv("paribas.csv", nrows = 20000)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
cols = ['float64', 'int64']
```

```python
numerical_columns = list(df.select_dtypes(include=cols))
```

```python
numerical_columns;
```

```python
df = df[numerical_columns]
```

```python
df.shape
```

```python
correlated_matrix = df.corr()
```

```python
correlated_matrix
```

```python
correlated_features = set()
```

```python
type(correlated_features)
```

```python
for i in range(len(correlated_matrix.columns)):
    for j in range(i):
        if abs(correlated_matrix.iloc[i, j]) > 0.8:
            colname = correlated_matrix.columns[i]
            correlated_features.add(colname)
```

```python
len(correlated_features)
```

```python
df.drop(correlated_features, axis = 1, inplace= True)
```

```python
df.shape
```

```python
"""
mpg -> target variable
take any 3 features and build the linear reg model
no need for cross validation
and evaluate -> calculate r2 score, mean absolute error
"""
```

```python
df = pd.read_csv("mtcars.csv")
df.head()
```

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x, y)
```

```python
lr.score(x, y)
```

### RFE

```python
rfe = RFE(LinearRegression(), n_features_to_select= 5)
```

```python
x = df.drop('mpg', axis = 1)
```

```python
rfe.fit(x, y)
```

```python
rfe.get_support()
```
