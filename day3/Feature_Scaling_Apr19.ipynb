```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

#### Normalization

```python
x = df[["Age", 'EstimatedSalary']]
```

```python
x
```

```python
plt.title("distibution of age and sal")
plt.xlabel("salary")
plt.ylabel("age")

plt.scatter(x["EstimatedSalary"], x["Age"])
plt.grid();
```

```python
x.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
plt.title("distibution of age and sal")
plt.xlabel("salary")
plt.ylabel("age")

plt.scatter(x_scaled[:, 1], x_scaled[:, 0])
plt.grid();
```
