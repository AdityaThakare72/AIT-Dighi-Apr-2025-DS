```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([12, 19, 23, 30, 37, 41, 50])
y = np.array([10, 20, 22, 30, 40, 47, 55])
z = np.array([53, 49, 41, 35, 30, 28, 20])
```

### Line plot

```python
plt.figure(figsize=(16, 9))

plt.title("change in y w.r.t x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.xticks(range(0, 55, 1))
plt.yticks(range(0, 60, 1))

plt.plot(x, y, marker = "o", mfc = "red", color = "black",
        ls = "--");
plt.grid()
# markers -> o, H, >, <, ^, *
```

```python
plt.figure(figsize=(16, 9))

plt.title("change in z w.r.t y x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.xticks(range(0, 55, 1))
plt.yticks(range(0, 60, 1))

plt.plot(z, y, marker = "o", mfc = "red", color = "purple",
        ls = "--");
plt.grid()
# markers -> o, H, >, <, ^, *
```

```python
plt.figure(figsize=(16, 9))

plt.title("change in variables")
plt.xlabel("x values")
plt.ylabel("y values")
plt.xticks(range(0, 55, 1))
plt.yticks(range(0, 60, 1))

plt.plot(x, y, marker = "o", mfc = "red", color = "black",
        ls = "--", label = "2024");
plt.plot(z, y, marker = "o", mfc = "red", color = "purple",
        ls = "--", label = "2025");

plt.grid()
plt.legend();
# markers -> o, H, >, <, ^, *
```

```python
# subplot
```

```python
plt.figure(figsize=(10, 9))

plt.subplot(2,2,2)
plt.title("change in y w.r.t x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x, y, marker = "o", mfc = "red", color = "black",
        ls = "--");
plt.grid()

plt.subplot(2,2,1)
plt.title("change in z w.r.t y x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(z, y, marker = "o", mfc = "red", color = "purple",
        ls = "--");
plt.grid()
```

### Bar Plot

```python
courses = ["Lua", "C#", "Dart", "Python", "Java"]
students = [45, 56, 49, 70, 68]
```

```python
plt.title("student enrollment")
plt.xlabel("courses")
plt.ylabel("no of students")

plt.bar(courses, students, color = 'black', edgecolor = "yellow",
       width = 0.3);
plt.grid()
# load the student.csv => plot bar graph to check which student got
# how many marks
```

```python
q3 = np.percentile(data["Age"], 75)
q1 = np.percentile(data["Age"], 25)
q3, q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + iqr * 1.5
lb = q1 - iqr * 1.5
ub, lb
```

```python
data[(data["Age"] > lb) & (data["Age"] < ub)]
```
